### 4. Agrupar dados com base em uma coluna usando pandas

```
import numpy as np
import pandas as pd
df = spark.read.csv("/FileStore/tables/retail.csv", header=True, sep=',')
df = df.toPandas()
df.head()
df.groupby('invoice_no').agg(n=('stock_code', 'count')).reset_index()
```

### 5. Usar a função max em alguma coluna do DataFrame

```
import numpy as np
import pandas as pd
df = spark.read.csv("/FileStore/tables/retail.csv", header=True, sep=',', inferSchema=True)
df = df.toPandas()
df['invoice_date'].max()
df.head()
```
